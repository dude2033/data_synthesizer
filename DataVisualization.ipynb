```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

import seaborn as sns
```

```python
def Plotwith3param(data1,data2,data3):
    xdata1 = []
    xdata2 = []
    xdata3 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    for d3 in data3[0]:
        xdata3.append(float(d3.strip('%')))

    return xdata1, xdata2, xdata3
```

```python
def Plotwith2param(data1,data2):
    xdata1 = []
    xdata2 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    return xdata1, xdata2
```

```python
def plotbox(title,xlabel,ylabel,data):
    df = pd.DataFrame.from_dict(data)
    plt.figure(figsize=(15,8))
    sns.set_style("whitegrid",rc={'figure.facecolor':'white'})
    #sns.set(rc={'figure.facecolor':'white'})
    boxplot = sns.boxplot(data=df,  orient='h', width=0.5,palette="pastel",linewidth=0.8,flierprops={"marker": "o"},
        boxprops={"facecolor": (.4, .6, .8, .5)},
        medianprops={"color": "coral"},notch=True)
    boxplot.tick_params(labelsize=20)
    boxplot.axes.set_title(title, fontsize=30)
    boxplot.set_xlabel(xlabel, fontsize=25)
    boxplot.set_ylabel(ylabel, fontsize=28)
```

```python
withoutNans = pd.read_csv("SyntheticData/withoutNans.csv", header=None)
withNans = pd.read_csv("SyntheticData/withalltheNansnointerpolation.csv", header=None)
withhalfNans = pd.read_csv("SyntheticData/with50%Nansnointerpolation.csv", header=None)

data1,data2,data3 = Plotwith3param(withNans,withhalfNans,withoutNans)

dicdata = {"With Nans" :data1, "With 50% Nans" : data2, "Without Nans": data3}

plotbox('With Nans vs Without Nans vs With 50% Nans',"Evaluation Score in %","Comparison",dicdata)
  
```
