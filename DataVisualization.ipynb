```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

import seaborn as sns
```

```python
def Plotwith3param(data1,data2,data3):
    xdata1 = []
    xdata2 = []
    xdata3 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    for d3 in data3[0]:
        xdata3.append(float(d3.strip('%')))

    return xdata1, xdata2, xdata3
```

```python
def Plotwith2param(data1,data2):
    xdata1 = []
    xdata2 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    return xdata1, xdata2
```

```python
def plotbox(title,xlabel,data):
    df = pd.DataFrame.from_dict(data)
    plt.figure(figsize=(15,8))
    sns.set_style("whitegrid",rc={'figure.facecolor':'white'})
    #sns.set(rc={'figure.facecolor':'white'})
    boxplot = sns.boxplot(data=df,  orient='h', width=0.5,palette="pastel",linewidth=0.8,flierprops={"marker": "o"},
        boxprops={"facecolor": (.4, .6, .8, .5)},
        medianprops={"color": "coral"},notch=True)
    boxplot.tick_params(labelsize=20)
    boxplot.axes.set_title(title, fontsize=30)
    boxplot.set_xlabel(xlabel, fontsize=25)
```

```python
withoutNans = pd.read_csv("SyntheticData/0%NaNs_notInterpolated_8Features_1542InputLines.csv", header=None)
withNans = pd.read_csv("SyntheticData/100%NaNs_notInterpolated_8Features_1542InputLines.csv", header=None)
withhalfNans = pd.read_csv("SyntheticData/50%NaNs_notInterpolated_8Features_1542InputLines.csv", header=None)

data1,data2,data3 = Plotwith3param(withNans,withhalfNans,withoutNans)

dicdata = {"0% NaNs" :data1, "50% NaNs" : data2, "100% NaNs": data3}

plotbox('0% NaNs vs 50% NaNs vs 100% NaNs \n  No Interpolation with N=1542 ',"Evaluation Score in %",dicdata)
  
```

```python
intall = pd.read_csv("SyntheticData/100%NaNs_Interpolated_8Features_1542InputLines.csv", header=None)
int50 = pd.read_csv("SyntheticData/50%NaNs_Interpolated_8Features_1542InputLines.csv", header=None)

data1,data2 = Plotwith2param(intall,int50)

dicdata = {"100% NaNs" : data1, "50% NaNs":data2}

plotbox("50% NaNs vs 100% NaNs \n Interpolation with N=1542","Evaluation Score in %",dicdata)
```

```python
feature8 = pd.read_csv("SyntheticData/100%NaNs_Interpolated_8Features_1542InputLines.csv", header=None)
feature12 = pd.read_csv("SyntheticData/100%NaNs_Interpolated_12Features_1542InputLines.csv", header=None)
feature18 = pd.read_csv("SyntheticData/100%NaNs_Interpolated_18Features_1542InputLines.csv", header=None)

data1,data2,data3 = Plotwith3param(feature8,feature12,feature18)

dicdata = {"Features 8" :data1, "Features 12" : data2, "Features 18": data3}

plotbox("Features 8 vs Features 12 vs Features 18 \n Interpolation with N=1542","Evaluation Score in %",dicdata)
```
