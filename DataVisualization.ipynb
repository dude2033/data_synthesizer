```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
def Plotwith3param(data1,data2,data3):
    xdata1 = []
    xdata2 = []
    xdata3 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    for d3 in data3[0]:
        xdata3.append(float(d3.strip('%')))

    return xdata1, xdata2, xdata3
```

```python
def Plotwith2param(data1,data2):
    xdata1 = []
    xdata2 = []

    for d1 in data1[0]:
        xdata1.append(float(d1.strip('%')))

    for d2 in data2[0]:
        xdata2.append(float(d2.strip('%')))

    return xdata1, xdata2
```

```python
def plotbox(title, data):
    fig = plt.figure(figsize =(15, 7))
    ax = plt.subplot()
    ax.set_title(title)

    ax.boxplot(data.values(), vert = 0)
    ax.set_yticklabels(data.keys())
    ax.xaxis.set_major_formatter(plt.FormatStrFormatter('% .2f'))
    #ax.set_xlim(74, 86)
    ax.set_xlabel('Socore in %')
```

```python
withoutNans = pd.read_csv("SyntheticData/withoutNans.csv", header=None)
withNans = pd.read_csv("SyntheticData/withalltheNansnointerpolation.csv", header=None)
withhalfNans = pd.read_csv("SyntheticData/with50%Nansnointerpolation.csv", header=None)

data1,data2,data3 = Plotwith3param(withNans,withhalfNans,withoutNans)

dicdata = {"With Nans" :data1, "With 50% Nans" : data2, "Without Nans": data3}

plotbox('With Nans vs Without Nans vs With 50% Nans',dicdata)
  
```

```python
val10 = pd.read_csv("SyntheticData/10%Nans10%validationSet.csv", header=None)
val30 = pd.read_csv("SyntheticData/10%Nans30%validationSet.csv", header=None)
val60 = pd.read_csv("SyntheticData/10%Nans60%validationSet.csv", header=None)

data1,data2,data3 = Plotwith3param(val10,val30,val60)

dicdata = {"10% Validation" :data1, "30% Validation" : data2, "60% Validation": data3}

plotbox("10% Validation Set vs 30% Validation Set vs 60% Validation Set",dicdata)
  
```

```python
withoutNans = pd.read_csv("SyntheticData/withoutNans.csv", header=None)
withNans = pd.read_csv("SyntheticData/withalltheNansnointerpolation.csv", header=None)
withhalfNans = pd.read_csv("SyntheticData/with50%Nansnointerpolation.csv", header=None)

data1,data2,data3 = Plotwith3param(withNans,withhalfNans,withoutNans)

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()
ax.plot(data1, color='red')
ax.plot(data2, color='blue')
ax.plot(data3, color='green')

ax.xaxis.set_major_locator(plt.MaxNLocator(7))
ax.xaxis.set_major_formatter(plt.FormatStrFormatter('% .2f'))
ax.set_ylabel('Socore in %')
ax.set_title("With Nans vs Without Nans vs With 50% Nans")
```
